## Step 1: Install Required Packages

Run this cell to install dependencies (if not already installed).

````python
# Install required packages
import sys
import subprocess

def install_if_missing(package):
    try:
        __import__(package)
    except ImportError:
        print(f"Installing {package}...")
        subprocess.check_call([sys.executable, "-m", "pip", "install", "-q", package])

# Check and install packages
for pkg in ['numpy', 'matplotlib', 'seaborn', 'scipy', 'scikit-learn']:
    install_if_missing(pkg)

print("✓ All packages ready!")
````

## Step 2: Import Libraries and Configure Settings

````python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import FancyBboxPatch, Circle, FancyArrowPatch
import warnings
warnings.filterwarnings('ignore')

# Configure for Nature Computational Science
plt.rcParams.update({
    'font.family': 'sans-serif',
    'font.sans-serif': ['Helvetica', 'Arial', 'DejaVu Sans'],
    'font.size': 6,
    'axes.labelsize': 7,
    'axes.titlesize': 7,
    'xtick.labelsize': 6,
    'ytick.labelsize': 6,
    'legend.fontsize': 6,
    'axes.linewidth': 0.5,
    'pdf.fonttype': 42,
    'figure.dpi': 100,
})

# Define colors for operator regimes
COLORS = {
    'stable': '#2E7D32',        # Green
    'plastic': '#FFA726',       # Orange  
    'unstable': '#E53935',      # Red
    'deeply_stable': '#1565C0'  # Blue
}

print("✓ Settings configured")
````

## Step 3: Generate Data

Creates synthetic hematopoiesis data with operator regimes.

````python
def generate_data(seed=42):
    """Generate synthetic hematopoiesis data."""
    np.random.seed(seed)
    
    # HSC: stable cells at origin
    hsc = np.random.randn(200, 2) * 0.5 + [-4, 0]
    hsc_regime = ['stable'] * 200
    hsc_type = ['HSC'] * 200
    
    # MPP: plastic/unstable at branch point
    mpp = np.random.randn(400, 2) * 0.7 + [-1, 0]
    mpp_regime = ['plastic'] * 240 + ['unstable'] * 160
    mpp_type = ['MPP'] * 400
    
    # Erythroid: mostly stable, some unstable/deeply stable
    ery = np.random.randn(700, 2) * 0.6 + [2, 2]
    ery_regime = ['stable'] * 490 + ['unstable'] * 140 + ['deeply_stable'] * 70
    ery_type = ['Erythroid'] * 700
    
    # Neutrophil: mostly stable, some unstable/deeply stable
    neu = np.random.randn(700, 2) * 0.6 + [2, -2]
    neu_regime = ['stable'] * 490 + ['unstable'] * 140 + ['deeply_stable'] * 70
    neu_type = ['Neutrophil'] * 700
    
    # Combine
    coords = np.vstack([hsc, mpp, ery, neu])
    regimes = np.array(hsc_regime + mpp_regime + ery_regime + neu_regime)
    cell_types = np.array(hsc_type + mpp_type + ery_type + neu_type)
    
    return coords, regimes, cell_types

coords, regimes, cell_types = generate_data()
print(f"✓ Generated {len(coords)} cells")
print(f"  Cell types: {', '.join(np.unique(cell_types))}")
print(f"  Regimes: {', '.join(np.unique(regimes))}")
````

## Step 4: Define Plotting Functions

Three functions for the three panels of Figure 7.

````python
def plot_panel_a(ax):
    """Panel A: Conceptual illustration."""
    ax.set_xlim(-0.5, 10.5)
    ax.set_ylim(-0.5, 6.5)
    ax.axis('off')
    
    # Title
    ax.text(5, 6.2, 'Operator-defined states distinguish dynamical properties',
            ha='center', va='top', fontsize=7)
    
    # Left: Expression-defined
    ax.text(2.5, 5.5, 'Expression-defined', ha='center', fontsize=7, fontweight='bold')
    ax.text(2.5, 5.15, '(traditional)', ha='center', fontsize=5.5, style='italic')
    
    box1 = FancyBboxPatch((0.5, 2.5), 4, 2.3, boxstyle="round,pad=0.1",
                          edgecolor='black', facecolor='#f5f5f5', linewidth=0.8)
    ax.add_patch(box1)
    
    # Same-colored cells
    for x, y in [(1.5, 3.5), (2.0, 4.0), (2.5, 3.2), (3.0, 4.2),
                 (3.5, 3.7), (4.0, 3.3), (1.8, 4.3), (3.8, 4.5)]:
        ax.add_patch(Circle((x, y), 0.15, color='#757575', alpha=0.7))
    
    ax.text(2.5, 2.65, 'Same cell type', ha='center', fontsize=5.5, style='italic')
    
    # Arrow
    arrow = FancyArrowPatch((4.8, 3.6), (5.7, 3.6), arrowstyle='->',
                           mutation_scale=15, linewidth=1.5, color='black')
    ax.add_patch(arrow)
    
    # Right: Operator-defined
    ax.text(7.5, 5.5, 'Operator-defined', ha='center', fontsize=7, fontweight='bold')
    ax.text(7.5, 5.15, '(dynamical)', ha='center', fontsize=5.5, style='italic')
    
    box2 = FancyBboxPatch((5.5, 2.5), 4, 2.3, boxstyle="round,pad=0.1",
                          edgecolor='black', facecolor='#f5f5f5', linewidth=0.8)
    ax.add_patch(box2)
    
    # Different-colored cells (different regimes)
    cells = [((6.5, 3.5), 'stable'), ((7.0, 4.0), 'plastic'),
             ((7.5, 3.2), 'stable'), ((8.0, 4.2), 'unstable'),
             ((8.5, 3.7), 'plastic'), ((9.0, 3.3), 'stable'),
             ((6.8, 4.3), 'unstable'), ((8.8, 4.5), 'deeply_stable')]
    for (x, y), regime in cells:
        ax.add_patch(Circle((x, y), 0.15, color=COLORS[regime], alpha=0.8))
    
    ax.text(7.5, 2.65, 'Different operator regimes', ha='center', fontsize=5.5, style='italic')
    
    # Legend
    for i, (regime, color) in enumerate(COLORS.items()):
        x = 1.0 + i * 2.2
        ax.add_patch(Circle((x, 1.5), 0.12, color=color, alpha=0.8))
        ax.text(x + 0.25, 1.5, regime.replace('_', ' ').title(),
               ha='left', va='center', fontsize=5.5)
    
    # Eigenvalue labels
    ax.text(1.5, 0.8, r'$\lambda_{\mathrm{max}} < 0$', ha='center', fontsize=5.5)
    ax.text(3.7, 0.8, r'$\lambda_{\mathrm{max}} \approx 0$', ha='center', fontsize=5.5)
    ax.text(5.9, 0.8, r'$\lambda_{\mathrm{max}} > 0$', ha='center', fontsize=5.5)
    ax.text(8.1, 0.8, r'$\lambda_{\mathrm{min}} \ll 0$', ha='center', fontsize=5.5)


def plot_panel_b(ax, coords, regimes, cell_types):
    """Panel B: UMAP with operator regimes."""
    for regime in ['stable', 'plastic', 'unstable', 'deeply_stable']:
        mask = regimes == regime
        ax.scatter(coords[mask, 0], coords[mask, 1], c=COLORS[regime],
                  s=4, alpha=0.6, label=regime.replace('_', ' ').title(),
                  rasterized=True, linewidths=0)
    
    ax.set_xlabel('UMAP 1', fontsize=7)
    ax.set_ylabel('UMAP 2', fontsize=7)
    ax.set_title('Hematopoiesis: Operator regimes', fontsize=7, pad=8)
    
    legend = ax.legend(loc='upper right', frameon=True, framealpha=0.95,
                      edgecolor='black', fontsize=5.5, markerscale=2.0)
    legend.get_frame().set_linewidth(0.5)
    
    # Cell type labels
    labels = {'HSC': [-4, 0], 'MPP': [-1, 0],
              'Erythroid': [2, 2], 'Neutrophil': [2, -2]}
    for name, pos in labels.items():
        ax.text(pos[0], pos[1], name, ha='center', va='center', fontsize=6,
               bbox=dict(boxstyle='round,pad=0.35', facecolor='white',
                        edgecolor='black', alpha=0.9, linewidth=0.5))
    
    ax.set_xticks([])
    ax.set_yticks([])
    for spine in ax.spines.values():
        spine.set_visible(False)


def plot_panel_c(ax, regimes, cell_types):
    """Panel C: Regime distribution by cell type."""
    types = ['HSC', 'MPP', 'Erythroid', 'Neutrophil']
    regime_order = ['stable', 'plastic', 'unstable', 'deeply_stable']
    
    # Calculate percentages
    data = []
    for ct in types:
        mask = cell_types == ct
        ct_regimes = regimes[mask]
        fracs = [(ct_regimes == r).sum() / len(ct_regimes) * 100
                 for r in regime_order]
        data.append(fracs)
    data = np.array(data)
    
    # Stacked bar chart
    x = np.arange(len(types))
    bottom = np.zeros(len(types))
    for i, regime in enumerate(regime_order):
        ax.bar(x, data[:, i], 0.65, bottom=bottom,
              label=regime.replace('_', ' ').title(),
              color=COLORS[regime], alpha=0.9,
              linewidth=0.5, edgecolor='white')
        bottom += data[:, i]
    
    ax.set_ylabel('Percentage (%)', fontsize=7)
    ax.set_xlabel('Cell type', fontsize=7)
    ax.set_title('Operator regime heterogeneity', fontsize=7, pad=8)
    ax.set_xticks(x)
    ax.set_xticklabels(types, fontsize=6)
    ax.set_ylim(0, 100)
    ax.set_yticks([0, 25, 50, 75, 100])
    
    legend = ax.legend(loc='upper right', frameon=True, framealpha=0.95,
                      edgecolor='black', fontsize=5.5)
    legend.get_frame().set_linewidth(0.5)
    
    ax.grid(axis='y', alpha=0.3, linestyle='--', linewidth=0.5)
    ax.set_axisbelow(True)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_linewidth(0.5)
    ax.spines['bottom'].set_linewidth(0.5)

print("✓ Plotting functions defined")
````

## Step 5: Create Complete Figure 7

Combines all three panels into the final figure.

````python
# Create figure (Nature CS max width: 180 mm = 7.09 inches)
fig = plt.figure(figsize=(7.09, 2.5))
gs = fig.add_gridspec(1, 3, wspace=0.35, left=0.02, right=0.98, top=0.88, bottom=0.12)

# Panel A
ax_a = fig.add_subplot(gs[0, 0])
plot_panel_a(ax_a)
ax_a.text(-0.8, 6.5, 'a', fontsize=8, fontweight='bold', va='top')

# Panel B
ax_b = fig.add_subplot(gs[0, 1])
plot_panel_b(ax_b, coords, regimes, cell_types)
ax_b.text(-5.8, 4.8, 'b', fontsize=8, fontweight='bold', va='top')

# Panel C
ax_c = fig.add_subplot(gs[0, 2])
plot_panel_c(ax_c, regimes, cell_types)
ax_c.text(-0.95, 108, 'c', fontsize=8, fontweight='bold', va='top')

# Title
fig.suptitle('Operator-defined cellular states capture dynamical stability regimes',
            fontsize=10, y=0.97)

plt.show()
print("✓ Figure 7 created!")
````

## Step 6: Save High-Resolution Outputs

````python
# Save at 600 DPI (final submission)
fig.savefig('figure7_600dpi.png', dpi=600, bbox_inches='tight',
            facecolor='white', pad_inches=0.05)
print("✓ Saved: figure7_600dpi.png (600 DPI)")

# Save at 300 DPI (review)
fig.savefig('figure7_300dpi.png', dpi=300, bbox_inches='tight',
            facecolor='white', pad_inches=0.05)
print("✓ Saved: figure7_300dpi.png (300 DPI)")

# Save as PDF (vector)
fig.savefig('figure7.pdf', bbox_inches='tight',
            facecolor='white', pad_inches=0.05)
print("✓ Saved: figure7.pdf (vector)")

print("\n" + "="*60)
print("SUCCESS! Figure 7 generated.")
print("="*60)
print("\nOutput files:")
print("  • figure7_600dpi.png - Final submission (600 DPI)")
print("  • figure7_300dpi.png - Review quality (300 DPI)")
print("  • figure7.pdf - Vector format")
print("\nNature Computational Science compliant:")
print("  ✓ Font: Helvetica (sans serif)")
print("  ✓ Text size: 5-7 pt")
print("  ✓ Panel labels: 8 pt bold")
print("  ✓ Width: 180 mm (7.09 inches)")
print("  ✓ Resolution: 600 DPI")
````

````python
import pandas as pd

print("\nOVERALL REGIME DISTRIBUTION:")
print("="*50)
regime_counts = pd.Series(regimes).value_counts()
for regime in ['stable', 'plastic', 'unstable', 'deeply_stable']:
    if regime in regime_counts.index:
        count = regime_counts[regime]
        pct = count / len(regimes) * 100
        print(f"{regime.replace('_', ' ').title():15s}: {count:4d} cells ({pct:5.1f}%)")

print("\nREGIME DISTRIBUTION BY CELL TYPE:")
print("="*50)
for ct in ['HSC', 'MPP', 'Erythroid', 'Neutrophil']:
    mask = cell_types == ct
    ct_regimes = regimes[mask]
    print(f"\n{ct}:")
    ct_counts = pd.Series(ct_regimes).value_counts()
    for regime in ['stable', 'plastic', 'unstable', 'deeply_stable']:
        if regime in ct_counts.index:
            count = ct_counts[regime]
            pct = count / len(ct_regimes) * 100
            print(f"  {regime.replace('_', ' ').title():15s}: {count:3d} cells ({pct:5.1f}%)")
````
