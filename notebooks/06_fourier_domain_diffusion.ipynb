# Fourier-Domain Diffusion Demo
This notebook illustrates k-space diffusion components added in the feature branch.

```python
import torch
from scIDiff.Fourier.transforms import dft, idft
from scIDiff.models.fourier_score_network import MultiBandScoreNet
x = torch.randn(4, 512)
X = dft(x)
net = MultiBandScoreNet(gene_dim=512)
s = net(X, t=0.5, cond={})
x_rec = idft(X)
x.shape, X.shape, s.shape, x_rec.shape
```
